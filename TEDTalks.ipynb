# 7. Count the total number of ratings received by each talk

```python
ratings_used = set()
```

```python
def ratings_count(value):
    count = 0
    for rating in value:
        ratings_used.add(rating['name'])
        count += rating['count']
    return count
```

```python
ted['num_ratings'] = ted['ratings_list'].apply(ratings_count)
```

```python
sns.histplot(ted[ted.num_ratings <= 10000],x='num_ratings')
```

```python
ratings_used
```

```python
negative_ratings = {'Confusing','Longwinded','Obnoxious','Unconvincing'}
```

```python
def percent_negative(ratings):
    
    count = negative_count = 0
    for rating in ratings:
        if rating['name'] in negative_ratings:
            negative_count += rating['count']
        count += rating['count']
    
    return round((negative_count/count) * 100,2)
            
            
```

```python
ted['percent_negative'] = ted.ratings_list.apply(percent_negative)
```

```python
def get_funny_ratings(group):
    
    
    funny_count = 0
    total_ratings = 0
    for ratings in group.ratings_list:
        for rating in ratings:
            total_ratings += rating['count']
            if rating['name'] == 'Funny':
                funny_count += rating['count']

    return round(funny_count/total_ratings * 100,2)
    
    
```

```python
occupation_counts = ted.speaker_occupation.value_counts()
```

```python
indexes = occupation_counts[occupation_counts >= 5].index
```

```python
from collections import Counter
```

```python
def most_frequent_rating(value):
    c = Counter()
    for rating in value:
        c[rating['name']] += rating['count']
    
    
    return c.most_common(1)[0][0]
        
```

```python
ted['most_common_rating'] = ted.ratings_list.apply(most_frequent_rating)
```
